```python
import pandas as pd
```

```python
corplist = pd.read_csv("./data/corplist-refine.csv")
indutycodesort = corplist.sort_values(by=['induty_code']).reset_index(drop = True)
indutycodesort = indutycodesort.astype('str')
```

```python
# 'intuty_code' 앞 2자리 추출
indutycodesort['induty_code_short'] = indutycodesort['induty_code'].str[:2]
```

```python
indutycodesort = indutycodesort.drop(['corp_name', 'induty_code'], axis=1)
```

```python
'''
금융업 (0)
    64 ~ 66.금융 및 보험업
IT정보통신업 (1)
    58 ~ 63.정보통신업
    26.전자 부품, 컴퓨터, 영상, 음향 및 통신장비 제조업
건설업 (2)
    41~ 42.건설업
    68.부동산업
    72.건축 기술, 엔지니어링 및 기타 과학기술 서비스업
화학/제약 (3)
    19.코크스, 연탄 및 석유정제품 제조업
    20.화학 물질 및 화학제품 제조업; 의약품 제외
    21.의료용 물질 및 의약품 제조업
    22.고무 및 플라스틱제품 제조업
음식료업 (4)
    10.식료품 제조업
    11.음료 제조업
    12.담배 제조업
    56.음식점 및 주점업
기계/장비 (5)
    27.의료, 정밀, 광학 기기 및 시계 제조업
    28.전기장비 제조업
    29.기타 기계 및 장비 제조업
    30.자동차 및 트레일러 제조업
    31.기타 운송장비 제조업
판매/유통 (6)
    45 ~ 47.도매 및 소매업
기타 (7)
'''
```

```python
# 한국표준산업분류표 코드 매칭 
def get_category(induty_code_short):
    cat = ''
    if induty_code_short in ["64", "65", "66"] : cat = '0'
    elif induty_code_short in ["26", "58", "59", "60", "61", "62", "63"]: cat = '1'
    elif induty_code_short in ["41", "42", "68", "72"] : cat = '2'
    elif induty_code_short in ["19", "20", "21", "22"]: cat = '3'
    elif induty_code_short in ["10", "11", "12", "56"]: cat = '4'
    elif induty_code_short in ["27", "28", "29", "30", "31"]: cat = '5'
    elif induty_code_short in ["45", "46", "47"]: cat = '6'
    else : cat = '7'
    
    return cat
```

```python
indutycodeshort_copy = indutycodesort
```

```python
indutycodeshort_copy['main_indus_code'] = indutycodeshort_copy['induty_code_short'].apply(get_category)
```

```python
indutycodeshort_copy
```

```python
def get_category2(induty_code_short):
    cat = ''
    if induty_code_short == '0' : cat = '금융업'
    elif induty_code_short == '1' : cat = 'IT정보통신업'
    elif induty_code_short == '2' : cat = '건설업'
    elif induty_code_short == '3' : cat = '화학제약'
    elif induty_code_short == '4' : cat = '음식료업'
    elif induty_code_short == '5' : cat = '기계장비'
    elif induty_code_short == '6' : cat = '판매유통'
    else : cat = '기타'
    
    return cat
```

```python
indutycodeshort_copy['main_indus_name'] = indutycodeshort_copy['main_indus_code'].apply(get_category2)
```
