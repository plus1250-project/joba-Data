```python
# xml 파일 읽기
import xml.etree.ElementTree as ET
tree = ET.parse('./data/CORPCODE.xml')
root = tree.getroot()

#stock_code 가 있는 기업의 corp_code만 code_list에 추가
code_list=[]
for x in range(len(root)):    
    #root[0][0].tag = 'corp_code'
    #root[0][1].tag = 'corp_name'
    #root[0][2].tag = 'stock_code'
    
    # stock_code가 있다면
    if root[x][2].text != " ":
        code_list.append(root[x][0].text)
        
#stock_code 가 있는 기업의 개수 확인        
len(code_list)

# 기업개황 불러오기
import requests
def load_data(corp_code):
    API_KEY="API_KEY"
    url = 'https://opendart.fss.or.kr/api/company.json?crtfc_key='+API_KEY+'&corp_code='+corp_code
    r = requests.get(url)
    company_data = r.json()
    return company_data
    
# 진행상황 확인
from tqdm import notebook

company_condition_list = []
for corp_code in notebook.tqdm(code_list):
    company_dict = load_data(corp_code)
    company_condition_list.append(company_dict)
    
# 기업 개황 저장
import pickle
with open('company_condition_list.txt','wb') as f:
       pickle.dump(company_condition_list,f)
```
